```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

from google.colab import files
uploaded = files.upload()

image_path = '1.jpg'  # Ganti dengan nama gambar yang Anda unggah
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

ksize = (15, 15)  # Ukuran kernel Gaussian (boleh diubah sesuai kebutuhan)
sigma_x = 0  # Nilai sigma-x (boleh diubah sesuai kebutuhan)
gaussian_image = cv2.GaussianBlur(image, ksize, sigma_x)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Gambar Asli')

plt.subplot(1, 2, 2)
plt.imshow(gaussian_image)
plt.title('Hasil Filter Gaussian')

plt.show()
```

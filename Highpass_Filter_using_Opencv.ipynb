```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

from google.colab import files
uploaded = files.upload()

image_path = '1.jpg'  # Ubah dengan nama gambar yang Anda unggah
image = cv2.imread(image_path)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #Konversi gambar menjadi grayscale

#Menampilkan gambar original
plt.subplot(2, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Gambar Asli')

plt.subplots_adjust(hspace=0.5)

kernel = np.array([[-1, -1, -1], #Kernel untuk filter highpass
                  [-1, 8, -1],
                  [-1, -1, -1]])

highpass_img = cv2.filter2D(image_gray, -1, kernel)

#Menampilkan hasil HPF Image
plt.subplot(2, 2, 2)
plt.imshow(highpass_img, cmap='gray')
plt.title('HPF Image')

plt.show()
```
